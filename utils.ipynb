```python
import os
from os import listdir
from os.path import isfile, join

path = '/home/hossein/Documents/git_mining'
data_dir = join(path, 'data')
files = [join(data_dir, f) for f in listdir(data_dir) if isfile(join(data_dir, f))]
files.sort(reverse=True)
```

```python
import pandas as pd

df1 = pd.read_csv(files[0])
df1 = df1[['Summary', 'Issue key', 'Issue id', 'Issue Type', 'Status', 'Resolution', 'Description']]
df1.head()
```

```python
df2 = pd.read_csv(files[1])
df2 = df2[['Summary', 'Issue key', 'Issue id', 'Issue Type', 'Status', 'Resolution', 'Description']]
df2.head()
```

```python
df3 = pd.read_csv(files[2])
df3 = df3[['Summary', 'Issue key', 'Issue id', 'Issue Type', 'Status', 'Resolution', 'Description']]
df3.head()
```

```python
df4 = pd.read_csv(files[3])
df4 = df4[['Summary', 'Issue key', 'Issue id', 'Issue Type', 'Status', 'Resolution', 'Description']]
df4.head()
```

```python
df = pd.concat([df1, df2, df3, df4], ignore_index=True)
df.head()
```

```python
df.info
```

```python
df.to_csv(join(data_dir, 'issues.csv'), index=False)
```
