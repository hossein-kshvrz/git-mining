```python
import os
from os import listdir
from os.path import isfile, join

path = '/home/hossein/Documents/git_mining'
data_dir = join(path, 'data')
files = [join(data_dir, f) for f in listdir(data_dir) if isfile(join(data_dir, f))]
files.sort(reverse=True)
```

```python
import pandas as pd

df1 = pd.read_csv(files[0])
df1 = df1[['Summary', 'Issue key', 'Issue id', 'Issue Type', 'Status', 'Resolution', 'Description']]
df1.head()
```

```python
df2 = pd.read_csv(files[1])
df2 = df2[['Summary', 'Issue key', 'Issue id', 'Issue Type', 'Status', 'Resolution', 'Description']]
df2.head()
```

```python
df3 = pd.read_csv(files[2])
df3 = df3[['Summary', 'Issue key', 'Issue id', 'Issue Type', 'Status', 'Resolution', 'Description']]
df3.head()
```

```python
df4 = pd.read_csv(files[3])
df4 = df4[['Summary', 'Issue key', 'Issue id', 'Issue Type', 'Status', 'Resolution', 'Description']]
df4.head()
```

```python
df = pd.concat([df1, df2, df3, df4], ignore_index=True)
df.head()
```

```python
df_min = df[['Issue key', 'SHA']]
df_min.head()
```

```python
import requests
import time

texts = []
keys = []

for index, row in df_min.iterrows():
    key = row['Issue key']
    sha = row['SHA']
    headers = {'content-type': 'application/json',
                   'accept': 'application/vnd.github.v3.diff'}
    url = 'https://github.com/apache/activemq/commit/'
    with open(join(path, 'auth.conf'), 'r') as file:
        lines = file.readlines()
    username = lines[0].split('\n')[0]
    password = lines[1].split('\n')[0]
    session = requests.Session()
    session.auth = (username, password)
    response = session.get(url + sha, headers=headers)
    if response.text not in texts:
        keys.append(key)
        texts.append(response.text)
    else:
        print(key, sha)
    time.sleep(2)
```
